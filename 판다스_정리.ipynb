```python
# 판다스 불러오기
import pandas as pd
```

```python
# 딕셔너리 형태로 만드는 방법
df = pd.DataFrame(
    {"a" : [4, 5, 6],
     "b" : [7, 8, 9],
     "c" : [10, 11, 12]},
    index = [1, 2, 3])

df
# 판다스는 데이터 전처리를 위한 라이브러리
```

```python
# 리스트 형태로 만드는 방법
df = pd.DataFrame(
    [[4,7,10],
     [5,8,11],
     [6,9,12]],
     index = [1, 2, 3],
     columns = ['a', 'b', 'c'])

df
```

```python
# 별책값(?) 넣어서 데이터 프레임 만들어보기
# 딕셔너리 형태
import numpy as np

df = pd.DataFrame(
    {"a" : [4, 5, 6, 6, np.nan],
     "b" : [7, 8, np.nan, 9, 9],
     "c" : [10, 11, 12, np.nan, 12]},
    index = pd.MultiIndex.from_tuples(
        [('d',1), ('d',2), ('e',2), ('e',3), ('e',4)],
    name = ['n', 'v']))

df
```

```python
df[df['a'] < 7]
# 논리 연산자로 사용했을 때는 불리언으로 나옴
# 결측지는 크기를 알 수 없으므로 False
```

```python
df[df['c'] >= 7]
```
